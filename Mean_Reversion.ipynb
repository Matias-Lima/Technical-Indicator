```python
import pandas as pd
import pandas_ta as ta
import yfinance as yf
import matplotlib.pyplot as plt
import talib as tl 
plt.style.use('default')

start_date = "2019-10-27"
end_date = "2022-08-02"
stock1 = "AAPL"
df = yf.download([stock1], start = start_date, end = end_date)
#print('data fields downloaded:', set(df.columns.get_level_values(0)))

#---------------- Mean Reversion --------------------
#Stochastic Slow (Momentum Indicators)
df['SLOW1'],df['SLOW2']= tl.STOCH(df['High'],df['Low'],df['Close'],1,5,0,5,0)
df[['Close','SLOW1','SLOW2']].plot(figsize=(12,12))
plt.show()

#Stochastic Fast (Momentum Indicators)
df['FAST1'],df['FAST2']= tl.STOCHF(df['High'],df['Low'],df['Close'],5,3)
df[['Close','FAST1','FAST2']].plot(figsize=(12,12))
plt.show()
#--------------------------------------------
#help(tl.get_functions())
#help(tl.STOCHF)
```
